```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from skimage import io
DISASTER='santa-rosa-wildfire'
IMAGEID='00000030'
XBD='/home/catskills/Desktop/dataxv2/xBD'
pre=f'{XBD}/{DISASTER}/images/{DISASTER}_{IMAGEID}_pre_disaster.png'
post=f'{XBD}/{DISASTER}/images/{DISASTER}_{IMAGEID}_post_disaster.png'
preimg = io.imread(pre, as_gray=True)
postimg = io.imread(post, as_gray=True)
x_width=preimg.shape[0]
window=5
center_chip=preimg[window:x_width-window]
chip_width=center_chip.shape[0]
W=list(range(-window,window))
plt.plot(W,[np.linalg.norm(postimg[window+i:chip_width+i+window]-center_chip) for i in W])
plt.savefig('SRLLC7.jpg')
```
